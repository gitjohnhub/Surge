```python
def readFiles(filename):
    rule =[]
    with open(filename) as file:
        lines = file.readlines()
        for line in lines:
            if not line.startswith('#'):
                rule.append(line)
        return rule
libyrule = readFiles("Liby.list")
nobydarule = readFiles("AdRule.list")
for rule in libyrule:
    if rule not in nobydarule:
        print(rule)
        
    
```
